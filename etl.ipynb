voy a traer el historial de los ultimos 4 años para cada moneda, cada moneda la guardare en un dataframe distinto

```python
def extraccion_historial(moneda:str):
    df = cg.get_coin_market_chart_by_id(id = moneda, vs_currency= 'usd', days= '1460') # el equivalente en dias de 4 años
    df= pd.DataFrame(data=df) 

    for i in df.keys():
    
        df[['fecha', i]] = pd.DataFrame(df[i].tolist(), columns=['fecha', i]) #separar la fecha de la data
    
    # la fecha es devuelta en formato unix, asi que con la libreria datetime se transformara la columna fecha de unix a
    # datetime con el siguiente bucle
    for i in df.index:
        df['fecha'][i] = df['fecha'][i] / 1000 # debido a que la fecha trae milisegundos se tiene que dividir para eliminar esos ceros
        df['fecha'][i] = datetime.datetime.fromtimestamp(df['fecha'][i])
        df['fecha'][i] =df['fecha'][i].strftime("%Y-%m-%d")  #solo se requiere el año, mes y dia
    
    return df
```

```python
dfs = []
for coin in coins:
    df = extraccion_historial(coin)
    dfs.append(df)
```
